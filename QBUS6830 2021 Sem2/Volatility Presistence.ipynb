```python
import numpy as np
import matplotlib.pyplot as plt
```

$$\mathbb{E}[\sigma_{t+1}^2|\mathcal{F}_{t-1}]=0.02 + 0.9\mathbb{E}[\sigma_{t}^2|\mathcal{F}_{t-1}]$$

$$\mathbb{E}[\sigma_{t+1}^2|\mathcal{F}_{t-1}]=0.02 + 0.7\mathbb{E}[\sigma_{t}^2|\mathcal{F}_{t-1}]$$

Note, in the Risk Metric Model we actually have
$$\mathbb{E}[\sigma_{t+1}^2|\mathcal{F}_{t-1}]=\mathbb{E}[\sigma_{t}^2|\mathcal{F}_{t-1}]$$

```python
N = 100
std0 = 0.03

model1, model2, model3 = np.zeros(N), np.zeros(N), np.zeros(N)

for i in range(N):
    if i == 0:
        model1[0], model2[0], model3[0] = std0, std0, std0
    else:
        model1[i] = 0.02 + 0.9*model1[i-1]
        model2[i] = 0.02 + 0.8*model2[i-1]
        model3[i] = model3[i-1]
```

```python
plt.figure(figsize=(15, 5))

plt.plot(model1, label='0.9 persistence')
plt.plot(model2, label='0.8 persistence')
plt.plot(model3, label='Risk Metric')

plt.plot(np.arange(N), np.ones(N)*0.2, '--', label='0.2 - long run volatility')
plt.plot(np.arange(N), np.ones(N)*0.1, '--', label='0.1 - long run volatility')

plt.legend();

plt.title('$\mathbb{E}[\sigma_{t+n}^2 | \mathcal{F}_{t-1}]$', fontsize=15);
plt.xlabel('n', fontsize=13);
plt.ylabel('$\sigma^2$', fontsize=13);
```
